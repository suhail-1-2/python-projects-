```python
correlation = df[['Confirmed',	'Deaths',	'Recovered',	'Active']].corr()
correlation
plt.figure(figsize=(7,4))
sns.heatmap(correlation)
```

```python
#in order to find the fetality rate we need to divide death to confirmed rate  
df['fetality_rate'] = df['Deaths']/ df['Confirmed'] *100
#in order to find the recovery rate we need to divide recovered to confirmed rate  
df['recovered_rate'] = df['Recovered']/ df['Confirmed'] *100
```

```python
plt.figure(figsize = (7,4))
sns.scatterplot(x = 'fetality_rate', y = 'recovered_rate', data=df)
plt.title('CASE FATALITY VS RECOVERY')
plt.xlabel('fetality')
plt.ylabel('recovery')
plt.show()
```
